**Special graphs: K3,3 and K5 - Smallests non-planar graph, according to Kuratowski**
# Complete Bipartite graph k3,3
*For me, **K‚ÇÉ,‚ÇÉ models** the incidence structure of Pappus‚Äô configuration ‚Äî in projective geometry:*

Pappus' theorem says:
Given two lines with three points on each (A, B, C on one line and A‚Ä≤, B‚Ä≤, C‚Ä≤ on the other), the intersection points of line pairs:

AB' ‚à© A'B

AC' ‚à© A'C

BC' ‚à© B'C

Lie on the same line ‚Äî **Are collinear**.

Now, returning to the graph, that is, A, B, C and A', B', C' are nodes or vertex and each line connecting these nodes are edges.

That's exactly how **K‚ÇÉ,‚ÇÉ** is structured:
Every point from one group connects to every point from the other ‚Äî complete bipartite structure.

*It cannot be drawn in a plane without edge crossings ‚Üí this is why **K‚ÇÉ,‚ÇÉ is not planar**.üèÆ
*Number of edges for a bipartite graph = m X n*

---



```python
import networkx as nx
import matplotlib.pyplot as plt

# --- K3,3: Complete Bipartite Graph with sets of size 3 and 3 ---
k33 = nx.complete_bipartite_graph(3, 3)

# Relabel nodes to represent input/output to resemble neuronal networks :)
# Reminds me Pappus' perpendicularity theorem.
label_map = {
    0: 'I', 1: 'II', 2: 'III',
    3: 'IV', 4: 'V', 5: 'VI'
}
k33 = nx.relabel_nodes(k33, label_map)  # relabel the same graph

# Position input nodes (top) and output nodes (bottom)
pos = {}
for i, node in enumerate(['I', 'II', 'III']):
    pos[node] = (i, 1)
for i, node in enumerate(['IV', 'V', 'VI']):
    pos[node] = (i, 0)


# Draw the graph
plt.figure(figsize=(6, 4))
nx.draw(k33,
        pos,
        with_labels=True,
        node_color='gold',
        node_size=800,
        edge_color='gray',
        width=2)
plt.title("K3,3 - Fully Connected Neural Layer (3‚Üí3)")
plt.axis('off')
plt.show()
```

**Example:**
*Let's say a transportation planer is creating a heritage walk for some customers and while developing a connection scheme between 7 cities he identifies a configuration within this network that's equivalent to **k3,3**. So what are the implications on constructing tracks without level crossing?*
*That means that walk is non-planar, implying that some tracks will require overpasses or underpasses to avoid some intersections.*

**K5 ‚Äî Complete graph with 5 vertex, 10 edges**

*We know that complete graph ùêæùëõ is Eulerian if and only if n%2==1 (i.e., ùëõ is odd) and ùëõ‚â•3.*

*Since each vertex in a complete graph ùêæùëõ  has a degree of ùëõ‚àí1, the degree is even when ùëõ is odd, and odd when ùëõ is even.*

*Moreover, in an Eulerian graph, we traverse every edge exactly once, starting and ending at the same vertex. In contrast, complete graphs are Hamiltonian, meaning they contain a Hamiltonian circuit ‚Äî a cycle that visits every vertex exactly once and also returns to the starting point*




```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# --- K5: Complete Graph with 5 nodes ---
k5 = nx.complete_graph(5)

# Get the layout
pos = nx.circular_layout(k5)

# Rotate the layout 90 degrees counterclockwise
rotated_pos = {node: np.array([-y, x]) for node, (x, y) in pos.items()}

plt.figure(figsize=(4, 4))
nx.draw(k5,
        pos=rotated_pos,
        with_labels=True,
        node_color='darkgreen',
        edge_color='gray',
        width=2 ,
        node_size=800,
        font_color='white')
plt.title("K5 - Complete Graph (Rotated 90¬∞)", pad=20)
plt.axis('off')
plt.tight_layout()
plt.show()
```
