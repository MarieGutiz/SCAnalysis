**Special graphs: K3,3 and K5 - Smallests non-planar graph, according to Kuratowski**
# Complete Bipartite graph k3,3
*For me, **Kâ‚ƒ,â‚ƒ models** the incidence structure of Pappusâ€™ configuration â€” in projective geometry:*

Pappus' theorem says:
Given two lines with three points on each (A, B, C on one line and Aâ€², Bâ€², Câ€² on the other), the intersection points of line pairs:

AB' âˆ© A'B

AC' âˆ© A'C

BC' âˆ© B'C

Lie on the same line â€” **Are collinear**.

Now, returning to the graph, that is, A, B, C and A', B', C' are nodes or vertex and each line connecting these nodes are edges.

That's exactly how **Kâ‚ƒ,â‚ƒ** is structured:
Every point from one group connects to every point from the other â€” complete bipartite structure.

*It cannot be drawn in a plane without edge crossings â†’ this is why **Kâ‚ƒ,â‚ƒ is not planar**.ğŸ®
*Number of edges for a bipartite graph = m X n*

---



```python
import networkx as nx
import matplotlib.pyplot as plt

# --- K3,3: Complete Bipartite Graph with sets of size 3 and 3 ---
k33 = nx.complete_bipartite_graph(3, 3)

# Relabel nodes to represent input/output to resemble neuronal networks :)
# Reminds me Pappus' perpendicularity theorem.
label_map = {
    0: 'I', 1: 'II', 2: 'III',
    3: 'IV', 4: 'V', 5: 'VI'
}
k33 = nx.relabel_nodes(k33, label_map)  # relabel the same graph

# Position input nodes (top) and output nodes (bottom)
pos = {}
for i, node in enumerate(['I', 'II', 'III']):
    pos[node] = (i, 1)
for i, node in enumerate(['IV', 'V', 'VI']):
    pos[node] = (i, 0)


# Draw the graph
plt.figure(figsize=(6, 4))
nx.draw(k33,
        pos,
        with_labels=True,
        node_color='gold',
        node_size=800,
        edge_color='gray',
        width=2)
plt.title("K3,3 - Fully Connected Neural Layer (3â†’3)")
plt.axis('off')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt


# Define raw products
baguettes = ['Bgt1', 'Bgt2', 'Bgt3','Bgt4']
cheeses = ['Chs1', 'Chs2','Chs3','Chs4','Chs5']
mortadellas = ['Mdl1', 'Mdl2','Mdl3']
toppings = ['Tpg1', 'Tpg2', 'Tpg3', 'Tpg4', 'Tpg5', 'Tpg6', 'Tpg7']


layers = [baguettes, cheeses, mortadellas, toppings]

# Create a directed graph, cos it makes sense :)
Sussys_net_pdt = nx.DiGraph()

# Add nodes and connect layers
for i in range(len(layers) - 1):
    for u in layers[i]:
        for v in layers[i + 1]:
            Sussys_net_pdt.add_edge(u, v)

# Compute positions for layout (layered)
pos = {}
x_spacing = 2
for layer_idx, layer in enumerate(layers):
    y_offset = len(layer) / 2
    for j, node in enumerate(layer):
        pos[node] = (layer_idx * x_spacing, y_offset - j)

# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(Sussys_net_pdt,
        pos,
        with_labels=True,
        node_color='yellow',
        node_size=1200,
        edge_color='blue',
        width=1.5,
        font_size=12)
plt.title("Sussy's Product Combinations: Baguette â†’ Cheese â†’ Mortadella â†’ Topping")
plt.show()
```

**And the question arrives, in the real-world analogy: Can Sussy organizes a product combinations without unavoidable overlap or "interchanging products" in the sense of network congestion???.**

**Finding ğ¾3,3 requires subgraph isomorphism.**ğŸ”„ğŸ˜

```python
import networkx as nx

# Define K3,3
K33 = nx.complete_bipartite_graph(3, 3)

#Recreate the graph, with the above info from Sussy's net
G = nx.Graph()
for i in range(len(layers) - 1):
    for u in layers[i]:
        for v in layers[i + 1]:
            G.add_edge(u, v)

# Subgraph isomorphism search
from networkx.algorithms import isomorphism

GM = isomorphism.GraphMatcher(G, K33)#Use GraphMatcher to check isomporphism
if GM.subgraph_is_isomorphic():
    print("Yes, this graph contains a K3,3 subgraph.")
else:
    print("No K3,3 subgraph found.")
```

Let's find a ğ¾3,3 inside Sussy's sandwich graph, then highlight its nodes and edges in a different color. ğŸ¥ª

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import isomorphism

# --- Subgraph isomorphism search ---
GM = isomorphism.GraphMatcher(G, K33)
found = GM.subgraph_is_isomorphic()

if found:
    mapping = next(GM.subgraph_isomorphisms_iter())
    k33_nodes = set(mapping.keys())
    k33_edges = [(u, v) for u, v in G.edges() if u in k33_nodes and v in k33_nodes]
    print("K3,3 subgraph found with nodes:", k33_nodes)
else:
    print("No K3,3 subgraph found.")
    k33_nodes, k33_edges = set(), []

# --- Compute layered layout ---
pos = {}
x_spacing = 2
for layer_idx, layer in enumerate(layers):
    y_offset = len(layer) / 2
    for j, node in enumerate(layer):
        pos[node] = (layer_idx * x_spacing, y_offset - j)

# --- Draw full graph ---
plt.figure(figsize=(12, 7))
nx.draw(G, pos,
        with_labels=True,
        node_color="lightgray",
        node_size=1200,
        edge_color="lightblue",
        width=1.5,
        font_size=10)

# --- Highlight K3,3 subgraph ---
if found:
    nx.draw_networkx_nodes(G, pos,
                           nodelist=k33_nodes,
                           node_color="orange",
                           node_size=1500)
    nx.draw_networkx_edges(G, pos,
                           edgelist=k33_edges,
                           edge_color="red",
                           width=2.5)

plt.title("Sussy's Product Graph with Highlighted K3,3 Subgraph", fontsize=14)
plt.axis("off")
plt.show()
```

**K5 â€” Complete graph with 5 vertex, 10 edges**

*We know that complete graph ğ¾ğ‘› is Eulerian if and only if n%2==1 (i.e., ğ‘› is odd) and ğ‘›â‰¥3.*

*Since each vertex in a complete graph ğ¾ğ‘›  has a degree of ğ‘›âˆ’1, the degree is even when ğ‘› is odd, and odd when ğ‘› is even.*

*Moreover, in an Eulerian graph, we traverse every edge exactly once, starting and ending at the same vertex. In contrast, complete graphs are Hamiltonian, meaning they contain a Hamiltonian circuit â€” a cycle that visits every vertex exactly once and also returns to the starting point*




```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# --- K5: Complete Graph with 5 nodes ---
k5 = nx.complete_graph(5)

# Get the layout
pos = nx.circular_layout(k5)

# Rotate the layout 90 degrees counterclockwise
rotated_pos = {node: np.array([-y, x]) for node, (x, y) in pos.items()}

plt.figure(figsize=(4, 4))
nx.draw(k5,
        pos=rotated_pos,
        with_labels=True,
        node_color='darkgreen',
        edge_color='gray',
        width=2 ,
        node_size=800,
        font_color='white')
plt.title("K5 - Complete Graph (Rotated 90Â°)", pad=20)
plt.axis('off')
plt.tight_layout()
plt.show()
```

For a complete graph the number of edges are given by the formula:
*n*(*n*+1)/2, where n are number of nodes, perhaps reminds you the *n*(*n*+1)/2 formula (the triangular formula) , that's the sum a series of number up to n, starting from 1, e.g. n=5 => **(5(6))/2= 15**. More formally,

$T_n = \sum_{i=1}^{n} i = \frac{n(n+1)}{2}$
, for the triangular formula. And

$E(K_n) = \frac{n(n-1)}{2}$ , for the sum of edges in a K-graph,

So for our K5 n=5 nodes => **(5(4))/2 = 10** edges, and that's because we don't count edges connecting to a node itself. This is why complete graphs are very natural in combinatorics â€” counting all **unordered pairs** of distinct elements gives a balanced, â€œcalibratedâ€ structure.

We know that The number of ways to choose ğ‘˜ elements from ğ‘› elements is given by the binomial coefficient $ \binom{n}{k} $
:
$ \frac{n!}{k!(n-k)!} $, for combinatoric, no order implied.

For the K5:
$ \binom{5}{2} $ = $ \frac{5!}{2!(5-2)!} $ = 10,

 which matches the number of edges in ğ¾5. This shows that there are 10 possible ways to connect 5 nodes in pairs, illustrating he combinatorial structure of a complete graphâ€”every pair of vertices is connected by an edge. Additionally, since each of the 5 vertices in ğ¾5 has degree 4, the graph also illustrates the **handshaking lemma**, which states that in any graph at **least two** vertices must share the same degree and that the sum of all vertex degrees is always even.In ğ¾5, the sum of all degrees is 5Ã—4=20, which is exactly twice the number of edges (since 10Ã—2=20). Here, all vertices share the same degree, and the total degree sum is 20, fully reinforcing the handshaking lemma.

Combinatorial counting is fundamental in probability theory.  Whenever we perform such counting, it is often associated with a **random** variable that can be described by a x - probability distribution, in our case, **discrete**.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Provide the Eulerian circuit for K5
euler_graph = DirectedEulerianGraph()
for u, v in k5.edges():
    euler_graph.add_edge(u, v)
    euler_graph.add_edge(v, u)  # make it symmetric for undirected

circuit = euler_graph.find_eulerian_circuit()
print("Eulerian circuit for K5:", circuit)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib

def draw_graph_with_path(G, path, title="Path"):
    pos = nx.spring_layout(G, seed=42)  # fixed layout for consistency
    plt.figure(figsize=(8,6))

    # Draw base graph
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1500, font_size=12, edge_color='gray')

    # Create edges from path
    path_edges = list(zip(path, path[1:]))

    # Generate a colormap (new API)
    cmap = matplotlib.colormaps["tab20"]  # 20 distinct colors
    colors = [cmap(i % 20) for i in range(len(path_edges))]

    # Draw edges with different colors
    for edge, color in zip(path_edges, colors):
        nx.draw_networkx_edges(G, pos, edgelist=[edge], edge_color=[color], width=2)

    plt.title(title, fontsize=14)
    plt.show()

```

**Eulerian circuit** rule:

You must traverse every edge exactly once ~ Edges may NOT repeat.

You are allowed to visit the same vertex multiple times ~ Vertices may repeat.

In the Hierholzer's algo every undirected edge becomes two directed edges
(u â†’ v and v â†’ u).

Therefore:

The Eulerian circuit must use all 20 directed edges.

Thatâ€™s why your circuit has 21 nodes (20 edges + return).
