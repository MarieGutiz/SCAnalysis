# Complete Bipartite graph k3,3
*For me, **K‚ÇÉ,‚ÇÉ models** the incidence structure of Pappus‚Äô configuration ‚Äî in projective geometry:*

Pappus' theorem says:
Given two lines with three points on each (A, B, C on one line and A‚Ä≤, B‚Ä≤, C‚Ä≤ on the other), the intersection points of line pairs:

AB' ‚à© A'B

AC' ‚à© A'C

BC' ‚à© B'C

Lie on the same line ‚Äî **Are collinear**.

Now, returning to the graph, that is, A, B, C and A', B', C' are nodes or vertex and each line connecting these nodes are edges.

That's exactly how **K‚ÇÉ,‚ÇÉ** is structured:
Every point from one group connects to every point from the other ‚Äî complete bipartite structure.

*It cannot be drawn in a plane without edge crossings ‚Üí this is why **K‚ÇÉ,‚ÇÉ is not planar**.üèÆ
*Number of edges for a bipartite graph = m X n*

---



```python
import networkx as nx
import matplotlib.pyplot as plt

# --- K3,3: Complete Bipartite Graph with sets of size 3 and 3 ---
k33 = nx.complete_bipartite_graph(3, 3)

# Relabel nodes to represent input/output to resemble neuronal networks :)
# Reminds me Pappus' perpendicularity theorem.
label_map = {
    0: 'I', 1: 'II', 2: 'III',
    3: 'IV', 4: 'V', 5: 'VI'
}
k33 = nx.relabel_nodes(k33, label_map)  # relabel the same graph

# Position input nodes (top) and output nodes (bottom)
pos = {}
for i, node in enumerate(['I', 'II', 'III']):
    pos[node] = (i, 1)
for i, node in enumerate(['IV', 'V', 'VI']):
    pos[node] = (i, 0)


# Draw the graph
plt.figure(figsize=(6, 4))
nx.draw(k33,
        pos,
        with_labels=True,
        node_color='gold',
        node_size=800,
        edge_color='gray',
        width=2)
plt.title("K3,3 - Fully Connected Neural Layer (3‚Üí3)")
plt.axis('off')
plt.show()
```

In general, we can think of a complete bipartite structure as representing two teams ‚Äî of size M x N ‚Äî where no members are connected within their own group, but all are connected to members of the other group. A classic example is a Supplier-Consumer relationship: suppliers don't interact with each other, but they're each linked to multiple consumers.

**Take away: No connections exist within each set.**

 So in total she has 4 * 5 * 3 * 7 = 420 **unique products - the sum of all the combination of sandwich baggettes**.

```python
import itertools

# Define options with emoji labels
baguettes = ['ü•ñ Baguette A', 'ü•ñ Baguette B', 'ü•ñ Baguette C', 'ü•ñ Baguette D']
cheeses = ['üßÄ Cheese Œë', 'üßÄ Cheese Œí', 'üßÄ Cheese Œì', 'üßÄ Cheese Œî', 'üßÄ Cheese Œï']
mortadellas = ['ü•ì Mortadella 1', 'ü•ì Mortadella 2', 'ü•ì Mortadella 3']
toppings = ['üßÇ Tomato', 'üßÇ Lettuce', 'üßÇ Pickles', 'üßÇ Mustard', 'üßÇ Mayonnoise', 'üßÇ Ketchup', 'üßÇ Cucumber']

# Generate Cartesian product of all sandwich combinations for Sussy
all_combinations = list(itertools.product(baguettes, cheeses, mortadellas, toppings))

# Print all sandwich options
print(f"Total sandwich combinations: {len(all_combinations)}\n")

for i, combo in enumerate(all_combinations, 1):
    print(f"{i:2}. {' + '.join(combo)}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt


# Define raw products
baguettes = ['Bgt1', 'Bgt2', 'Bgt3','Bgt4']
cheeses = ['Chs1', 'Chs2','Chs3','Chs4','Chs5']
mortadellas = ['Mdl1', 'Mdl2','Mdl3']
toppings = ['Tpg1', 'Tpg2', 'Tpg3', 'Tpg4', 'Tpg5', 'Tpg6', 'Tpg7']


layers = [baguettes, cheeses, mortadellas, toppings]

# Create a directed graph, cos it makes sense :)
Sussys_net_pdt = nx.DiGraph()

# Add nodes and connect layers
for i in range(len(layers) - 1):
    for u in layers[i]:
        for v in layers[i + 1]:
            Sussys_net_pdt.add_edge(u, v)

# Compute positions for layout (layered)
pos = {}
x_spacing = 2
for layer_idx, layer in enumerate(layers):
    y_offset = len(layer) / 2
    for j, node in enumerate(layer):
        pos[node] = (layer_idx * x_spacing, y_offset - j)

# Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(Sussys_net_pdt,
        pos,
        with_labels=True,
        node_color='yellow',
        node_size=1200,
        edge_color='blue',
        width=1.5,
        font_size=12)
plt.title("Sussy's Product Combinations: Baguette ‚Üí Cheese ‚Üí Mortadella ‚Üí Topping")
plt.show()
```

**Let's check for planarity for Sussy' Fresh bread & Pastry :**ü§î

---



```python
is_planar, P = nx.check_planarity(Sussys_net_pdt)

if is_planar: print("This graph is planar. No K3,3 subgraph was found.")
else: print("This graph contains a K3,3 subgraph (non-planar by Kuratowski).")
```

**K5 ‚Äî Complete graph with 5 vertex, 10 edges**

*We know that complete graph ùêæùëõ is Eulerian if and only if n%2==1 (i.e., ùëõ is odd) and ùëõ‚â•3.*

*Since each vertex in a complete graph ùêæùëõ  has a degree of ùëõ‚àí1, the degree is even when ùëõ is odd, and odd when ùëõ is even.*

*Moreover, in an Eulerian graph, we traverse every edge exactly once, starting and ending at the same vertex. In contrast, complete graphs are Hamiltonian, meaning they contain a Hamiltonian circuit ‚Äî a cycle that visits every vertex exactly once and also returns to the starting point*




```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# --- K5: Complete Graph with 5 nodes ---
k5 = nx.complete_graph(5)

# Get the layout
pos = nx.circular_layout(k5)

# Rotate the layout 90 degrees counterclockwise
rotated_pos = {node: np.array([-y, x]) for node, (x, y) in pos.items()}

plt.figure(figsize=(4, 4))
nx.draw(k5,
        pos=rotated_pos,
        with_labels=True,
        node_color='darkgreen',
        edge_color='gray',
        width=2 ,
        node_size=800,
        font_color='white')
plt.title("K5 - Complete Graph (Rotated 90¬∞)", pad=20)
plt.axis('off')
plt.tight_layout()
plt.show()
```
