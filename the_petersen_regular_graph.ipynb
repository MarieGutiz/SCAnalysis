**The Petersen graph(regular - degree 3 - 10 nodes - 15 edges - non-planar üèÆ)**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the Petersen graph
G = nx.petersen_graph()

# Draw the graph
plt.figure(figsize=(6, 6))
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=800, font_size=14)
plt.title("Petersen Graph mesh")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Create the Petersen graph
# G = nx.petersen_graph() # Uncheck when needed

# Generate positions: 5 outer pentagon points, 5 inner star points
angle = np.linspace(0, 2 * np.pi, 6)[:-1]  # 5 points
outer_radius = 1
inner_radius = 0.5

# Outer pentagon nodes (0-4)
outer = {i: (outer_radius * np.cos(a), outer_radius * np.sin(a)) for i, a in enumerate(angle)}

# Inner star nodes (5-9), rotated to be between outer nodes
inner = {i + 5: (inner_radius * np.cos(a + np.pi/5), inner_radius * np.sin(a + np.pi/5)) for i, a in enumerate(angle)}

# Combine positions
pos = {**outer, **inner}

# Draw the graph
plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_color='lightcoral', edge_color='black',
        node_size=800, font_size=12, font_weight='bold')
plt.title("Petersen Graph (Pentagon + Star Layout)")
plt.axis('equal')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Create Petersen graph
# G = nx.petersen_graph()

# Parameters
outer_radius = 1
inner_radius = 0.5

# To align both tips up:
# - Outer pentagon tip at 90¬∞ (œÄ/2)
# - Inner star tip is 36¬∞ ahead ‚Üí rotate both by (œÄ/2 - 0) and (œÄ/2 - œÄ/5)
outer_rotation = np.pi / 2   # 90¬∞
inner_rotation = np.pi / 2   # 90¬∞

# Base angles for 5 points
angle = np.linspace(0, 2 * np.pi, 6)[:-1]

# Outer pentagon nodes (0‚Äì4)
outer = {
    i: (
        outer_radius * np.cos(a + outer_rotation),
        outer_radius * np.sin(a + outer_rotation)
    ) for i, a in enumerate(angle)
}

# Inner star nodes (5‚Äì9), shifted to place star tip at top
inner = {
    i + 5: (
        inner_radius * np.cos(a + inner_rotation),
        inner_radius * np.sin(a + inner_rotation)
    ) for i, a in enumerate(angle)
}

# Combine and draw
pos = {**outer, **inner}
plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_color='gold', edge_color='black',
        node_size=800, font_size=12, font_weight='bold')
plt.title("Petersen Graph (Star and Pentagon Tips Aligned watching top - Nicely done)")
plt.axis('equal')
plt.show()
```
